```python
#H/W 
plt.axvline(x=0.2, ymin=0, ymax=0.8, c='red');
plt.axvline(x=0.4, ymin=0.3, ymax=0.7, c='orange');
plt.axvline(x=0.6, ymin=0, ymax=0.6, c='green');
```

```python
#H/W 
plt.axhline(y=0.2, xmin=0, xmax=0.8, c='red');
plt.axhline(y=0.4, xmin=0.3, xmax=0.7, c='orange');
plt.axhline(y=0.6, xmin=0, xmax=0.6, c='green');
```

```python
#H/W 
plt.axvline(x=0.2, ymin=0, ymax=0.8, c='red');
plt.axvline(x=0.4, ymin=0.3, ymax=0.7, c='orange');
plt.axvline(x=0.6, ymin=0, ymax=0.6, c='green');

plt.axhline(y=0.2, xmin=0, xmax=0.8, c='red');
plt.axhline(y=0.4, xmin=0.3, xmax=0.7, c='orange');
plt.axhline(y=0.6, xmin=0, xmax=0.6, c='green');
```

```python
cords= [0.2,0.5,0.7]
colors=['red', 'green', 'blue']
for cord, color in zip(cords, colors):
    plt.axvline(x=cord, ymin=0, ymax=1, c=color)
    plt.axhline(y=cord, xmin=0, xmax=1, c=color)
```

```python
plt.axvline(x=0.2, ymin=0, ymax=1, c='green', label="Green Line")
plt.legend(loc=10)
plt.xlabel("X Axis Label")
plt.ylabel("Y Axis Label", c="red")
plt.title("Plot Title", color="red");
```

##### -vlines()
-hlines()

```python
plt.vlines(x=0.2, ymin=-1.8, ymax=2, color='k');
plt.hlines(y=0.2, xmin=-1.8, xmax=2, color='red');
```

```python
plt.vlines(x=0.2, ymin=-1.8, ymax=2, label="Black Line", color='k', linestyles="dashdot");
plt.hlines(y=0.2, xmin=-1.8, xmax=2, label="Red Line", color='red', linestyles="dotted");
plt.legend();
```
